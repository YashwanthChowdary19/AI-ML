# Arrays in python

- Array is a type of collection 
- Array is a data structure that stores the multiple values of same datatype(integer, float, characters) in a continuous block of memory
- Array is not a built-in datatype 
- Arrays can be used in python by importing the array module
- Arrays are fast compared to list
- Array is a mutable datastructure
