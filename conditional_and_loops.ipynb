# conditional statements


Conditional statements are programming statements that allow the program to make decisions based on conditions (true or false).

```python
# if : Used to execute a block of code only when the condition is true.
```

```python
age = 18
if age >= 18:
    print("You are eligible to vote")
```

```python
# if - else : 
- Used when there are two possibilities.

- Runs one block if the condition is true, otherwise runs another block.
```

```python
age = 17 
if age <= 18 : 
    print("You are eligible to vote")
else : 
    print("Not eligible")
```

```python
# if - elif - else 
- Used when you want to check multiple conditions one after another.
```

```python
score = 75
if score >= 90:
    print("A Grade")
elif score >= 75:
    print("B Grade")
else:
    print("C Grade")
```

```python
# Nested if : When an if statement is written inside another if. 
age = 45
if age >= 18 : 
    if age < 60 : 
        print("adult")
        
```

# Loops

- Loops are used in programming to repeat a block of code multiple times.
- Instead of writing the same code again and again, loops help to automate repetition.

```python
# for loop 
- Used when you know how many times you want to repeat something or when looping through a sequence (list, string, range, etc.)
```

```python
# syntax 
for variable in sequence : 
    // block of code
```

```python
for i in range(10) : 
    print(i,end=" ")
```

```python
# while Loop

- Used when you do not know how many times the loop should run and want to repeat until a condition becomes false.
```

```python
i = 1 
while i <= 5 : 
    print(i,end=" ")
    i+=1
```

```python
# for-else in Python

- The else block in a for loop runs only when the loop completes normally (i.e., without being interrupted by break).
```

```python
# syntax
for item in sequence:
    # loop code
else:
    # executes if loop finishes without break
```

```python
for i in range(5) : 
    print(i)
else : 
    print("for loop completed normally")
```

```python
# while-else 
- Similar to for-else, the else part executes only if the while loop finishes normally, not when stopped by break.
```

```python
while condition:
    # loop code
else:
    # executes if loop completes
```

```python
i = 1 
while i <= 5 : 
    print(i)
    i+=1
else : 
    print("loop completed successfully")
```

```python
# Loop control statements
- Loop control statements are used to change the normal flow of a loop (for or while).
- They help us control how and when loops should stop, continue, or skip certain parts.
```

```python
# break Statement

- Stops the loop immediately

- Control moves to the next statement after the loop
```

```python
for i in range(5) : 
    if i==3 : 
        break 
    print(i)
```

```python
# continue Statement

- Skips the current iteration of the loop

- Moves directly to the next iteration
```

```python
for i in range(5) : 
    if i==3 : 
        continue 
    print(i)
```

```python
# pass Statement

- Does nothing

- Used as a placeholder when a statement is needed syntactically but you donâ€™t want any action
```

```python
for i in range(5) : 
    pass
```

```python
"""
ðŸŽ¯ Mini Project: Student Marks Management System

This project will:

Take input for multiple students

Calculate total and average

Decide pass/fail using if-else

Use loops to enter multiple records

Use break, continue, pass

Use for-else to search a studentâ€™s record
"""
```

```python
students = []

while True:
    print("\n--- Student Marks Entry System ---")

    name = input("Enter student name: ")
    marks = int(input("Enter marks (0-100): "))

    # continue example - skip invalid marks
    if marks < 0 or marks > 100:
        print("Invalid marks! Try again.")
        continue

    # Store data
    students.append((name, marks))
    print("Record added successfully!")

    # Ask if user wants to continue
    choice = input("Do you want to add more students? (yes/no): ")
    if choice.lower() == "no":
        break  # break example
```

```python
print("\n--- Student Results ---")

for name, marks in students:
    if marks >= 35:
        result = "Pass"
    else:
        result = "Fail"
    print(f"{name} - {marks} - {result}")
```

```python
search_name = input("\nEnter name to search: ")

for name, marks in students:
    if name == search_name:
        print(f"Record Found: {name} scored {marks}")
        break
else:  # only if no break occurs
    print("Student not found!")
```
