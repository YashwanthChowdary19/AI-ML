# characteristics of dictionary

```python
# mutable 
py_dict = {1 : "one", 2 : "two", "three" : 3} 
print(py_dict)
py_dict[1] = "ONE"
print(py_dict)  # one changed to ONE
```

```python
# unordered  : (before Python 3.7, order was not guaranteed; from Python 3.7+, insertion order is preserved).
```

```python
# duplicate values but no duplicate keys 
py_dict = {"name" : "yashwanth", "age":20, "branch" : "CSE"}
print(py_dict)
py_dict["updated_age"] = 20 # new pair is added to the dict
print(py_dict)
py_dict["age"] = 21 # no pair added to dict but it modified the value of age
print(py_dict) 
```

```python
# accessing dictionary values using keys as index
py_dict = {"name" : "yashwanth", "age":20, "branch" : "CSE"} 
print(py_dict["name"])
print(py_dict["age"])
print(py_dict["branch"])
```

```python
student1 = {
    "name":"yashwanth", 
    "age":21, 
    "branch":"CSE", 
    "marks": {
        "python":100, 
        "Java":99, 
        "javascript":98
    },
    "cgpa":9.0
}
print(student1.get("marks"))
print(student1.get("marks").get("python"))
```

```python
# modifying the values in dictionary 
student1["marks"]["c++"] = 97
student1
```

```python
# deleting pair in dictionary 
del student1["marks"]["c++"]
```

```python
"name" in student1   # it always checks the keys not the values
```

```python
"yashwanth" in student1
```

```python
"yashwanth" in student1.values() # now it is checks in values
```

```python
"yashwanth" not in student1.values()
```

```python
# loops 
for key,value in student1.items() : 
    print(key,"---->",value)
```

# simple project on dictionary 

```python
""" 
ðŸ›’ Grocery Store Billing System (Using Dictionary)
Project Explanation

This project stores grocery item names with their prices inside a dictionary.
The user enters items they want to purchase and the quantity.
The program calculates the total bill.
"""
```

```python
# grocery items with prices 
current_items = {
    "Milk": 30,
    "Bread": 25,
    "Eggs": 6,
    "Rice": 60,
    "Oil": 120,
    "Sugar": 45
}

# user enter the items with comma 
customer_items = input("enter the present items with separated by comma : ")

customer_list = list(customer_items.split(",")) 

total_amount = 0 

for each_item in customer_list : 
    item = each_item.strip()   # remove extra spaces
    if item in current_items : 
        total_amount += current_items[item]
    else:
        print(item, "not available!")

print("customer items list : ", customer_list)
print("total amount : ", total_amount)
```
