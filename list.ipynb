# Characteristics of list

- Dynamic
- can contains hetergenoeus values 
- Index based 
- Ordered 
- Mutable 
- Supports slicing 
- Allows duplicate values 
- Supports nested lists
- Supports many built in functions 
- iterable

#### Dynamic 
- list size can grow and shrink automatically

```python
fruits = ["Apple", "Banana", "Orange"]
print(fruits)
fruits.append("Grapes")
print(fruits)
```

#### allows mixed values 
- list stores the mixed values


```python
variable = ["fruits","vegetables",1,10.20,True]
print(variable)
```

#### index based
- lists allows to access the values using index values

```python
fruits = ["Apple", "Banana", "Orange"]
print(fruits[0])
print(fruits[1])
print(fruits[2])
```

#### supports slicing 
- list allows to accessing the values from one index to another using slicing

```python
variable = ["fruits","vegetables",1,10.20,True]
print(variable[1:4])
```

#### ordered 
- Maintains the order of items as you insert them
- The order "Apple" ‚Üí "Banana" ‚Üí "Cherry" is preserved.

```python
fruits = ["Apple", "Banana", "Cherry"]
print(fruits)
```

#### mutable
- list allows to change the values after it is created 

```python
fruits = ["Apple", "Banana", "Orange"]
fruits[1] = "Mango" # Banana changed to Mango
print(fruits) 
```

#### allows duplicate values


```python
fruits = ["Apple", "Banana", "Orange", "Apple", "Banana", "Orange"]
print(fruits)
```

#### supports built-in functions

```python
# append()
fruits = ["Apple", "Banana", "Orange"]
fruits.append("Cherry")
print(fruits)
# insert() 
fruits.insert(2,"Kiwi")
print(fruits)
# clear() 
fruits.clear()
print(fruits)
```

#### iterable

```python
fruits = ["Apple", "Banana", "Orange", "Kiwi", "Cherry"]
for i in fruits : 
    print(i)
```

#### adding multiple values using slicing

```python
fruits = ["Apple", "Banana", "Orange"]
fruits[3:5] = ["Cherry", "Kiwi"]
print(fruits)
```

#### changing multiple values using slicing

```python
fruits = ["Apple", "Banana", "Orange"]
fruits[1:3] = ["Cherry", "Kiwi"]
print(fruits)
```

### Shallow Copy

A shallow copy creates a new object, but does not copy the inner elements.
Instead, it only copies the references of inner items.

‚û°Ô∏è So, if you change inner data in one list, it will affect the other.

Short Definition:

Shallow copy copies only the top-level structure, not nested objects.

```python
import copy 
a = [1, 2, 3, [4, 5, 6]]
b = copy.copy(a)
```

```python
print(a)
print(b)
```

```python
b[1]=20 
print(a)
print(b)
```

```python
b[3][1]=50 
print(a)
print(b)
```

### Deep Copy

A deep copy creates a new object and also copies all nested elements completely.
There are no shared references between copies.

‚û°Ô∏è Changing inner data in the new list does not affect the original.

Short Definition:

Deep copy copies everything including nested objects (full independent duplicate).

```python
import copy 
a = [1, 2, 3, [4, 5, 6]]
b = copy.deepcopy(a)
```

```python
print(a)
print(b)
```

```python
b[1] = 20 
print(a)
print(b)
```

```python
b[3][1] = 50 
print(a)
print(b)
```
