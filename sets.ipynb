# set operations

```python
# union (|)
set1 = {1,2,3,4}
set2 = {4,5,6,7}
set1 | set2
```

```python
# intersection (&) 
set1 & set2
```

```python
# difference (-)
set1 - set2
```

```python
set2 - set1
```

```python
# symmetric difference (^)
set1 ^ set2
```

```python
# membership 
set1 in set2
```

```python
set1 not in set2
```

```python
# loops 
for i in set1 : 
    print(i)
```

```python
# difference_update 
set1 = {1,2,3,4}
set2 = {4,6,7,8}
set1.difference_update(set2)
print(set1)
```

```python
# symmetric_difference_update 
set1 = {1,2,3,4}
set2 = {4,6,7,8}
set1.symmetric_difference_update(set2)
print(set1)
```

```python
customer1 = {"Milk", "Bread", "Eggs", "Butter"}
customer2 = {"Bread", "Eggs", "Rice", "Oil"}
customer3 = {"Eggs", "Bread", "Juice", "Butter"}

# items commonly purchased by all customers.
common_items = customer1.intersection(customer2, customer3)
print("Common items purchased by all customers:", common_items)

# total variety of items sold today (unique items overall)
total_variety_items = customer1.union(customer2, customer3)
print("Total variety of items sold today:", total_variety_items)

# Unique items bought by each customer.
print("Unique items bought by customer1:", customer1.difference(customer2, customer3))
print("Unique items bought by customer2:", customer2.difference(customer1, customer3))
print("Unique items bought by customer3:", customer3.difference(customer1, customer2))
```

```python
def read_customer_items(customer_name):
    raw = input(f"Enter items for {customer_name} separated by commas: ")
    # Normalize: strip whitespace and lowercase to avoid duplication errors
    items = {item.strip().lower() for item in raw.split(",") if item.strip()}
    return items

def main():
    print("Grocery Store: Unique Customer Items Tracker")
    
    # Read customers (example: 3 customers). Could be dynamic.
    c1 = read_customer_items("Customer 1")
    c2 = read_customer_items("Customer 2")
    c3 = read_customer_items("Customer 3")

    while True:
        print("\nMenu:")
        print("1. Items purchased by all customers (intersection)")
        print("2. All unique items sold today (union)")
        print("3. Items unique to a customer (difference)")
        print("4. Compare two customers (symmetric difference)")
        print("5. Show all data")
        print("6. Exit")
        choice = input("Choose (1-6): ").strip()

        if choice == "1":
            print("Common items:", c1.intersection(c2, c3))
        elif choice == "2":
            print("All unique items:", c1.union(c2, c3))
        elif choice == "3":
            which = input("Which customer (1/2/3)? ").strip()
            if which == "1":
                print("Unique to Customer 1:", c1.difference(c2, c3))
            elif which == "2":
                print("Unique to Customer 2:", c2.difference(c1, c3))
            elif which == "3":
                print("Unique to Customer 3:", c3.difference(c1, c2))
            else:
                print("Invalid customer")
        elif choice == "4":
            a = input("Compare which two customers? (e.g., 1 2): ").split()
            if len(a) >= 2:
                pair = {a[0].strip(), a[1].strip()}
                if pair == {"1", "2"}:
                    print("Different between c1 & c2:", c1.symmetric_difference(c2))
                elif pair == {"1", "3"}:
                    print("Different between c1 & c3:", c1.symmetric_difference(c3))
                elif pair == {"2", "3"}:
                    print("Different between c2 & c3:", c2.symmetric_difference(c3))
                else:
                    print("Invalid pair")
            else:
                print("Invalid input")
        elif choice == "5":
            print("Customer1:", c1)
            print("Customer2:", c2)
            print("Customer3:", c3)
        elif choice == "6":
            print("Goodbye!")
            break
        else:
            print("Invalid choice, try again.")

if __name__ == "__main__":
    main()
```
